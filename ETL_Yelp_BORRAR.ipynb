## Archivo Business

```python
# Abrimos el archivo con info de comercios:
business = pd.read_pickle("Datasets/business.pkl")
business.head()
```

```python
business.info()
```

```python
business["categories"]
```

```python
business = business.iloc[:,0:14]
business.info()
```

```python
df = business.copy()
df.head()
```

```python
df["categories"] = df["categories"].str.split(',')
df = df.explode("categories")
df.head()
```

```python
df["categories"].unique()
```

```python
# Se filtran los restaurantes
palabra_clave = "restaurant"
restaurantes = df.dropna(subset="categories")
filtro = restaurantes["categories"].str.contains(palabra_clave, case=False)
restaurantes = restaurantes[filtro]
restaurantes.head()
```

```python
# Se chequean los valores únicos
restaurantes["categories"].value_counts()
```

```python
atributos_restaurant = restaurantes[["business_id", "attributes"]]
atributos_restaurant = atributos_restaurant.reset_index().drop(columns="index")
atributos_restaurant = pd.concat([atributos_restaurant, pd.json_normalize(atributos_restaurant["attributes"])], axis=1) 
atributos_restaurant.head()
```

```python
atributos_restaurant.columns
```

## Archivo reviews

```python
chunk_size = 100000 
chunks = pd.read_json("review-001.json", lines=True, chunksize=chunk_size)

# Procesar cada chunk y luego concatenar los resultados
dfs = []
for chunk in chunks:
    dfs.append(chunk)


df_reviews = pd.concat(dfs, axis=0, ignore_index=True)
df_reviews.head()
```
