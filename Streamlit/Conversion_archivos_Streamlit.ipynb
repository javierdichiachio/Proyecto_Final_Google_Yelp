```python
import pandas as pd
```

# Google Maps

```python
# Se carga el archivo de restaurantes:
res_g = pd.read_parquet("Data/Restaurants_Google.parquet")
res_g.head()
```

```python
res_g.columns
```

```python
res_g.info()
```

```python
# Nos quedamos solamente con las columnas que vamos a utilizar:
res_g = res_g[['gmap_id','name','latitude', 'longitude']]
res_g
```

```python
# Se carga la tabla que contiene las ciudades
res2 = pd.read_csv("Data/restaurante_google_index.csv")
res2.rename(columns={"restaurante":"name"}, inplace=True)
```

```python
res_g_concat = pd.merge(res_g, res2[["name", "city"]], on="name", how="inner")
res_g_concat.head()
```

```python
# Se carga el df de reviews:
rev_g = pd.read_parquet("Data/Reviews_Google_Sentiment.parquet")
rev_g.head()
```

```python
rev_g.columns
```

```python
rev_g.info()
```

```python
# Nos quedamos solamente con las columnas que vamos a utilizar:
rev_g = rev_g[['gmap_id','rating_final']]
rev_g.head()
```

```python
# Se realiza el merge entre la tabla de restaurantes y reviews:
df_google = pd.merge(res_g_concat,rev_g, on="gmap_id", how="inner")
df_google.info()
```

```python
df_google.head()
```

```python
df_google.columns
```

```python
# Se agrupan los valores por restaurante:
df_google_final = df_google.groupby(['gmap_id', 'name', 'city', 'latitude', 'longitude'])['rating_final'].agg(['mean', 'count']).reset_index()
df_google_final.head()
```

```python
df_google_final.info()
```

```python
# Se fija el nombre de las columnas:
df_google_final.columns = ['id_restaurant', 'name','city','latitude', 'longitude','rating','num_of_reviews']
df_google_final.head()
```

# Yelp

```python
# Se carga el archivo de restaurantes:
res_y = pd.read_parquet("Data/Restaurants_Yelp.parquet")
res_y.head()
```

```python
res_y.columns
```

```python
res_y.info()
```

```python
# Nos quedamos solamente con las columnas que vamos a utilizar:
res_y = res_y[['business_id','name','city', 'latitude', 'longitude']]
res_y.head()
```

```python
# Se carga el df de reviews:
rev_y = pd.read_parquet("Data/Reviews_Yelp_Sentiment.parquet")
rev_y.head()
```

```python
rev_y.columns
```

```python
rev_y.info()
```

```python
# Nos quedamos solamente con las columnas que vamos a utilizar:
rev_y = rev_y[['business_id','rating']]
rev_y.head()
```

```python
# Se realiza el merge entre la tabla de restaurantes y reviews:
df_yelp = pd.merge(res_y,rev_y, on="business_id", how="inner")
df_yelp.info()
```

```python
df_yelp.head()
```

```python
df_yelp.columns
```

```python
# Se agrupan los valores por restaurante:
df_yelp_final = df_yelp.groupby(['business_id', 'name', 'city', 'latitude', 'longitude'])['rating'].agg(['mean', 'count']).reset_index()
df_yelp_final.head()
```

```python
df_yelp_final.info()
```

```python
# Se fija el nombre de las columnas:
df_yelp_final.columns = ['id_restaurant', 'name', 'city','latitude', 'longitude','rating','num_of_reviews']
df_yelp_final.head()
```

# UniÃ³n Google Maps y Yelp

```python
# Se unen verticalmente ambos dataframe:
df_union = pd.concat([df_google_final, df_yelp_final], ignore_index=True)
df_union.head()
```

```python
df_union.info()
```

```python
# Se agrega columna provisoria para controlar que no existan ubicaciones duplicadas:
df_union["control_ubi"] = df_union["latitude"] + df_union["longitude"]
df_union.head()
```

```python
# Se controla que no haya valores duplicados en la columna creada:
df_union["control_ubi"].duplicated().sum()
```

```python
# Se eliminan los valores duplicados y la columna de control:
df_google_yelp = df_union.drop_duplicates(subset="control_ubi").reset_index(drop=True)
df_google_yelp.drop(columns=['control_ubi','latitude', 'longitude'], inplace=True)
df_google_yelp.info()
```

```python
# Se redondea el valor de la columna "rating" a un decimal:
df_google_yelp["rating"] = df_google_yelp["rating"].round(1)
df_google_yelp.head()
```

```python
# Se carga la lista de categorias del restaurant:
categorias = pd.read_parquet("Data/Categories_restaurants.parquet")
categorias = categorias[["restaurant_id", "category"]]
categorias.head()
```

```python
# Se eliminan los valores duplicados
categorias = categorias.drop_duplicates()
```

```python
categorias.shape
```

```python
# Se agrupan las diferentes categorias en una lista:
categories_imploded = categorias.groupby('restaurant_id')["category"].agg(list).reset_index()
categories_imploded.head()
```

```python
categories_imploded["category"].value_counts()
```

```python
categories_imploded.shape
```

```python
# Se cambia el nombre de la columna id para poder hacer un merge con el dataframe de google_yelp:
categories_imploded.rename(columns={"restaurant_id":"id_restaurant"}, inplace=True)
```

```python
# Se unen las categorÃ­as al dataframe original:
df_google_yelp_final = pd.merge(df_google_yelp, categories_imploded, on="id_restaurant", how="inner").reset_index(drop=True)
df_google_yelp_final.head()
```

```python
df_google_yelp_final.info()
```

```python
# Se exporta el archivo en formato csv para ser utilizado en Streamlit:
df_google_yelp_final.to_csv("Restaurants_merged_Streamlit.csv", index=False)
```
