```python
merged2["stars"]=merged2["stars"].fillna(merged2["avg_rating"])
```

```python
merged2["name_y"]=merged2["name_y"].fillna(merged2["name_x"])
```

```python
merged2.isnull().sum()
```

```python
merged2.drop(["avg_rating","name_x"], axis=1, inplace=True)
```

```python
merged2.rename(columns={'stars': 'rating'}, inplace=True)
```

```python
merged2
```

```python
dummies = pd.get_dummies(merged2['Attributes'])
```

```python
dummies
```

```python
df = pd.concat([merged2, dummies], axis=1)
```

```python
df
```

```python
df.drop(["Attributes","Attribute_orig"], axis=1, inplace=True)   
```

```python
df
```

```python
input=["id_restaurant","name_y","rating"]
```

```python
input_restaurant = df[input]
```

```python
input_restaurant
```

```python
input_restaurant["rating"]
```

```python
#Producto escalar para obtener los pesos
userProfile = dummies.transpose().dot(input_restaurant['rating'])
#Perfil del usuario
userProfile
```

```python
#Ahora llevemos los gÃ©neros de cada pelÃ­cula al marco de datos original
genreTable = df.set_index(df['id_restaurant'])
```

```python
genreTable = genreTable.drop(columns=['id_restaurant',"name_y","rating"])
```

```python
genreTable
```

```python
genreTable.shape
```

```python
#Multiplicando los gÃ©neros por los pesos para luego calcular el peso promedio
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())

recommendationTable_df.head()
```

```python
#Ordena nuestra recomendaciÃ³n en orden descendente

recommendationTable_df = recommendationTable_df.sort_values(ascending=False)

#Miremos los valores

recommendationTable_df
```

```python
#Tabla de recomendaciones final
recomendaciones=df.loc[df['id_restaurant'].isin(recommendationTable_df.head().keys())]
```
