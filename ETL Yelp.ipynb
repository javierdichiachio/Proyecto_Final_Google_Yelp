```python
chunk_size = 100000 
chunks = pd.read_json("review-001.json", lines=True, chunksize=chunk_size)

# Procesar cada chunk y luego concatenar los resultados
dfs = []
for chunk in chunks:
    dfs.append(chunk)


df_reviews = pd.concat(dfs, axis=0, ignore_index=True)
df_reviews.head()
```
